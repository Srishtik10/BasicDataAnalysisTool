```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def load_data(file_path):
    """Load the CSV file into a Pandas DataFrame."""
    try:
        data = pd.read_csv(file_path)
        print("Data loaded successfully!\n")
        print(data.head())
        return data
    except FileNotFoundError:
        print("Error: File not found. Please check the file path.")
        return None

def calculate_statistics(data):
    """Calculate basic statistics: mean, median, and mode for numeric columns."""
    print("\nBasic Statistics:\n")
    for column in data.select_dtypes(include=np.number).columns:
        mean = np.mean(data[column])
        median = np.median(data[column])
        mode = data[column].mode().iloc[0]  # Mode might return multiple values, taking the first
        print(f"{column} - Mean: {mean:.2f}, Median: {median:.2f}, Mode: {mode:.2f}")

def visualize_data(data, column):
    """Visualize data for a specific column."""
    if column not in data.columns:
        print(f"Error: Column '{column}' not found in the dataset.")
        return

    plt.figure(figsize=(10, 6))
    plt.hist(data[column], bins=10, color='skyblue', edgecolor='black')
    plt.title(f"Distribution of {column}", fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.ylabel("Frequency", fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

def main():
    # Load the dataset
    file_path = "mtcars.csv"  # Ensure the file is in the same directory or provide a full path
    data = load_data(file_path)

    if data is not None:
        # Perform basic data analysis
        calculate_statistics(data)

        # Visualize results for a column
        column = input("\nEnter the column to visualize: ").strip()
        visualize_data(data, column)

# Run the program
if __name__ == "__main__":
    main()
```
